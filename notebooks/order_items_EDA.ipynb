```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file_path = r'C:\Users\Admin\Documents\GitHub\brazilian-ecommerce-capstone\data\raw\olist_order_items_dataset.csv'
df_items = pd.read_csv(file_path)
df_items.head()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df_items['price'], bins=30, range=(0,500))
plt.title('Distribution of Product Prices')
plt.xlabel('price')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df_items['price'], bins=100)
plt.title('Full Distribution of Product Prices')
plt.xlabel('price')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df_items['freight_value'],bins=30)
plt.title('Distribution of Freight Price')
plt.xlabel('freight_value')
plt.ylabel('frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df_items['price'], df_items['freight_value'], alpha=0.3)
plt.title('Product Price vs Freight Value')
plt.xlabel('Product Price (₹)')
plt.ylabel('Freight Value (₹)')
plt.grid(True)
plt.show()
```
